```python
# Filter the original dataframe and mapping enso labels
clustering_df = df[(df['year'] <= 2017) & (df['year'] >= 1890)]
clustering_df = clustering_df[clustering_df['year'].isin(enso_labels.keys())]
clustering_df['enso_label'] = clustering_df['year'].map(enso_labels)

# onehot encode ENSO phases
clustering_df = pd.get_dummies(clustering_df, columns=['enso_label'], prefix='enso')

# melt dataframe so each enso phase gets a row
melted = clustering_df.melt(
    id_vars=['latitude', 'longitude', 'temperature'],
    value_vars=['enso_E', 'enso_L', 'enso_N'],
    var_name='enso_phase',
    value_name='is_phase'
)

# filter to keep only the relevant enso phase rows
melted = melted[melted['is_phase'] == 1]

# groupby location and enso phase then pivot
pivot = melted.groupby(['latitude', 'longitude', 'enso_phase'])['temperature'] \
              .mean().unstack(fill_value=0).reset_index()

# standardize enso response columns
clustering_cols = ['enso_E', 'enso_L', 'enso_N']
X = pivot[clustering_cols]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# kmeans clustering
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10)
pivot['cluster'] = kmeans.fit_predict(X_scaled)

# plot colored by cluster
plt.figure(figsize=(10, 6))
plt.scatter(pivot['longitude'], pivot['latitude'], c=pivot['cluster'], cmap='tab10', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Locations Based on ENSO Temperature Response')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()

# average enso response per cluster
print(pivot.groupby('cluster')[clustering_cols].mean())
```
