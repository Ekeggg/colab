```python
from scipy.stats import zscore

# Z-scores, assuming this data has normal distribution
df['temp_zscore'] = zscore(df['temperature'])
# Define outliers as Z-score > 3 or < -3
outliers = df[(df['temp_zscore'] > 3) | (df['temp_zscore'] < -3)]
df = df.drop(columns=['temp_zscore'])
len(outliers)
```

```python
df['temp_smooth'] = df['temperature'].rolling(window=3, center=True).mean()
```

```python
df = df.dropna()
```

```python
new_df = df[(df['year'] > 1889) & (df['year']<2018)]
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans

sampled_df = new_df.sample(n=100000, random_state=10)
X = sampled_df[['latitude', 'longitude', 'month', 'temp_smooth']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
silhouette_scores = []
db_scores = []
k_values = range(2, 10)  # k=2 to k=9

for k in k_values:
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=0, n_init=10, batch_size=1000)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled[:20000], cluster_labels[:20000])
    db_score = davies_bouldin_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    db_scores.append(db_score)

# plot silhouette and DB scores
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Silhouette Score', color=color)
ax1.plot(k_values, silhouette_scores, marker='o', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # second y-axis (db)
color = 'tab:red'
ax2.set_ylabel('Davies-Bouldin Score', color=color)
ax2.plot(k_values, db_scores, marker='x', color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Silhouette and Davies-Bouldin Scores vs Number of Clusters')
plt.show()

```

This graph is a graph of the number of clusters, and the silhouette and DB scores. This shows that even at the most optimal number of clusters (6), and even after seasonal smoothing, there seems to be a decent amount of noise in the data.
